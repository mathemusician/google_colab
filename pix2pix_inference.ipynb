```
%cd /content/

def delete_real(dir):
  files = os.listdir(dir)

  for file_name in files:
    if 'real' in file_name:
      os.remove(os.path.join(dir, file_name))

def rename_fake(dir):
  files = os.listdir(dir)

  for file_name in files:
    if 'fake' in file_name:
      new_file_name = file_name.split('_')[0] + '.png'
      os.rename(os.path.join(dir, file_name), os.path.join(dir, new_file_name))

delete_real(os.path.join('images', 'test'))
delete_real(os.path.join('images', 'train'))
delete_real(os.path.join('images', 'valid'))

rename_fake(os.path.join('images', 'test'))
rename_fake(os.path.join('images', 'train'))
rename_fake(os.path.join('images', 'valid'))
```
